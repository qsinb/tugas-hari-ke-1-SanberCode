<h2>Soal 2: Aritmatika di Python </h2>
    
- Buat statement pertambahan antara 2 number di Python
- Buat statement perkurangan antara 2 number di Python
- Buat statement perkalian antara 2 number di Python
- Buat statement pembagian antara 2 number di Python

```python
#pertambahan
a1 = 5
a2 = 5
total = a1 + a2 
print("total pertambahan = ", total)

#perkurangan
a1 = 5000
a2 = 1000
total = a1 - a2
print("total pengurangan = ", total)

#perkalian
a1 = 10
a2 = 6
total = a1 * a2
print("total perkalian = ", total)

#pembagian
a1 = 16
a2 = 4
total = a1 / a2
print("total pembagian = ", total)
```
